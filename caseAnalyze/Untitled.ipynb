```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.linear_model import LinearRegression
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm
from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.tools.tools import add_constant
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
# read the data of increae case
case_by_date_data= pd.read_csv("../data/increase_case_by_date.csv")
#read the data of stock index
stock_by_date_data = pd.DataFrame( pd.read_csv("../data/stock_shanghai.csv"),columns=["date","close","volume"])
#convert the format of date and  increase
case_by_date_data["date"] ="0"+ case_by_date_data["date"]+"20"
case_by_date_data["date"]
stock_by_date_data["stock_increase"] = 'null'
case_by_date_data
```

```python
for i in range(0,len(stock_by_date_data)):
    if i == 0:
        stock_by_date_data['stock_increase'][i] = 0
    else:
        
        stock_by_date_data['stock_increase'][i] = float((stock_by_date_data['close'][i] - stock_by_date_data['close'][i-1])/stock_by_date_data['close'][i-1])
```

```python
case_by_date_data["gobal_c_increase"] = 'null'
for i in range(0,len(case_by_date_data)):
    if i == 0:
        case_by_date_data["gobal_c_increase"][i] = 0
    else:
        case_by_date_data["gobal_c_increase"][i]  = float((case_by_date_data["confirmed_cases_global"][i] - case_by_date_data["confirmed_cases_global"][i-1])/case_by_date_data["confirmed_cases_global"][i-1])
        
        
case_by_date_data["gobal_r_increase"] = 'null'
for i in range(0,len(case_by_date_data)):
    if i == 0:
        case_by_date_data["gobal_r_increase"][i] = 0
    else:
        case_by_date_data["gobal_r_increase"][i]  = float((case_by_date_data["recovered_cases_global"][i] - case_by_date_data["recovered_cases_global"][i-1])/case_by_date_data["recovered_cases_global"][i-1])  
        

case_by_date_data["gobal_d_increase"] = 'null'
for i in range(0,len(case_by_date_data)):
    if i == 0:
        case_by_date_data["gobal_d_increase"][i] = 0
    else:
        case_by_date_data["gobal_d_increase"][i]  = float((case_by_date_data["dead_cases_global"][i] - case_by_date_data["dead_cases_global"][i-1])/case_by_date_data["dead_cases_global"][i-1])
        
        
        
        
```

```python
case_by_date_data["china_c_increase"] = 'null'
for i in range(0,len(case_by_date_data)):
    if i == 0:
        case_by_date_data["china_c_increase"][i] = 0
    else:
        case_by_date_data["china_c_increase"][i]  = float((case_by_date_data["confirmed_cases_China"][i] - case_by_date_data["confirmed_cases_China"][i-1])/case_by_date_data["confirmed_cases_China"][i-1])



case_by_date_data["china_r_increase"] = 'null'
for i in range(0,len(case_by_date_data)):
    if i == 0:
        case_by_date_data["china_r_increase"][i] = 0
    else:
        case_by_date_data["china_r_increase"][i]  = float((case_by_date_data["recovered_cases_China"][i] - case_by_date_data["recovered_cases_China"][i-1])/case_by_date_data["recovered_cases_China"][i-1])
        
case_by_date_data["china_d_increase"] = 'null'
for i in range(0,len(case_by_date_data)):
    if i == 0 or case_by_date_data["dead_cases_China"][i-1] ==0 :
        case_by_date_data["china_d_increase"][i] = 0
    else:
        case_by_date_data["china_d_increase"][i]  = float((case_by_date_data["dead_cases_China"][i] - case_by_date_data["dead_cases_China"][i-1])/case_by_date_data["dead_cases_China"][i-1])

        
```

```python
# convert to datetime
stock_by_date_data["date"] = pd.to_datetime(stock_by_date_data["date"],format="%Y-%m-%d")
case_by_date_data["date"] = pd.to_datetime(case_by_date_data["date"],format="%m/%d/%Y")
```

```python
stock_by_date_data["stock_increase"] = stock_by_date_data.stock_increase.astype(float)
case_by_date_data["gobal_c_increase"] = case_by_date_data.gobal_c_increase.astype(float)
case_by_date_data["gobal_d_increase"] = case_by_date_data.gobal_d_increase.astype(float)
case_by_date_data["gobal_r_increase"] = case_by_date_data.gobal_r_increase.astype(float)
case_by_date_data["china_c_increase"] = case_by_date_data.china_c_increase.astype(float)
case_by_date_data["china_d_increase"] = case_by_date_data.china_d_increase.astype(float)
case_by_date_data["china_r_increase"] = case_by_date_data.china_r_increase.astype(float)
```

```python
# create a new table 
merged_data = pd.merge(stock_by_date_data,case_by_date_data,on="date")

merged_data
```

```python
# anova
formula = 'stock_increase~ gobal_c_increase + gobal_r_increase + gobal_d_increase + china_c_increase + china_r_increase + china_d_increase'
anova_results = anova_lm(ols(formula,merged_data).fit())
anova_results
merged_data.iloc[:,4:-1]
```

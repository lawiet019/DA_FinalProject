```python
import pandas as pd 
```

```python
csv_data = pd.read_csv("../data/DXYArea.csv")
```

```python
#add a new line named date
csv_data["updateTime"]  = pd.to_datetime(csv_data["updateTime"],format='%Y-%m-%d %H:%M:%S')
csv_data["date"] = csv_data["updateTime"].dt.date
```

```python
csv_global_comfirmed = pd.read_csv("../data/time_series_covid19_confirmed_global.csv")
csv_global_dead = pd.read_csv("../data/time_series_covid19_deaths_global.csv")
csv_global_recovered = pd.read_csv("../data/time_series_covid19_recovered_global.csv")

csv_global_dead.head(6)
```

```python
csv_China_comfirmed = csv_global_comfirmed[csv_global_comfirmed["Country/Region"] =="China"]
csv_China_dead = csv_global_dead[csv_global_dead["Country/Region"] =="China"]
csv_China_recovered = csv_global_recovered[csv_global_recovered["Country/Region"] =="China"]
```

```python

date_columns = csv_global_comfirmed.columns.values[4:]
sum_confirmed_of_date = []
sum_dead_of_date = []
sum_recovered_of_date = []
sum_China_comfirmed = []
sum_China_dead = []
sum_China_recovered=[]
for cur_date in date_columns:
    sum_confirmed_of_date.append(csv_global_comfirmed[cur_date].sum())
    sum_dead_of_date.append(csv_global_dead[cur_date].sum())
    sum_recovered_of_date.append(csv_global_recovered[cur_date].sum())
    sum_China_comfirmed.append(csv_China_comfirmed[cur_date].sum())
    sum_China_dead.append(csv_China_dead[cur_date].sum())
    sum_China_recovered.append(csv_China_recovered[cur_date].sum())
   
   
    


    
    
    
    
    
```

```python
# create new dataframe
global_data = {
    "date":date_columns,
    "confirmed_cases_global":sum_confirmed_of_date,
    "dead_cases_global":sum_dead_of_date,
    "recovered_cases_global":sum_recovered_of_date,
    "confirmed_cases_China":sum_China_comfirmed,
    "dead_cases_China":sum_China_dead,
    "recovered_cases_China":sum_China_recovered
}

df_global_data = pd.DataFrame(global_data)
df_global_data
```
